```python
df_ventas_totales=pd.read_csv('VENTAS_TOTALES.csv',sep=';')
```

```python
# QuÃ© tipo de datos tengo
```

```python
df_ventas_totales.dtypes
```

```python
#Normalizo con dtos string poniendolos en minÃºscula
df_ventas_totales.SEGMENTO=df_ventas_totales.SEGMENTO.str.lower()
df_ventas_totales.NOMBREPRODUCTO=df_ventas_totales.NOMBREPRODUCTO.str.lower()
df_ventas_totales.TIPOVENTA=df_ventas_totales.TIPOVENTA.str.lower()
```

```python
df_ventas_totales
```

```python
# Borro los registros NAN
```

```python
df_ventas_totales=df_ventas_totales.dropna()
```

```python
df_ventas_totales
```

```python
# Recalculo campo Importe Ventas
```

```python
df_ventas_totales.IMPORTEVENTAS=df_ventas_totales['CANTIDADUNIDADES'].mul(df_ventas_totales['PRECIOVENTA'])
```

```python
df_ventas_totales
```

```python
# Borro la columna Fecha Ventas.
```

```python
df_ventas_totales=df_ventas_totales.drop(columns='FECHAVENTA')
```

```python
df_ventas_totales
```

```python
#genero dataframe prductos con 3 columnas y renombro columnas
df_productos=df_ventas_totales[['PRODUCTO','NOMBREPRODUCTO','IMPORTEVENTAS']]
df_productos.rename(columns={'NOMBREPRODUCTO':'NOMBRE','IMPORTEVENTAS':'TOTAL DE VENTAS'},inplace=True)
```

```python
df_productos
```

```python
#agrupo las tres columnas por el producto y su nombre y luego calculo la suma total de cada producto. 
df_productos=df_productos.groupby(['PRODUCTO','NOMBRE']).sum()
```

```python
df_productos
```

```python
#genero del archivo.csv productos
df_productos.to_csv('PRODUCTOS_piaacuna.csv',sep=';')
```
